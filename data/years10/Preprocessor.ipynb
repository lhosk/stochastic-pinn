```python
import pandas as pd
import glob
import numpy as np

# files = sorted(glob.glob("/content/drive/MyDrive/FinalProject/sPINN//data/years10/*.csv"))  # adjust path if needed
# df_list = [pd.read_csv(f) for f in files]
# df = pd.concat(df_list, ignore_index=True)

# turn csv into df
df = pd.read_csv("/content/drive/MyDrive/FinalProject/sPINN/data/years10/51214.csv")
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df["datetime"] = pd.to_datetime(
    df["YY"].astype(str) + "-" +
    df["MM"].astype(str) + "-" +
    df["DD"].astype(str) + " " +
    df["hh"].astype(str) + ":" +
    df["mm"].astype(str)
)
```

```python
df["hour"] = df["datetime"].dt.hour + df["datetime"].dt.minute/60
df["time_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
```

```python
other_cols = ["time_sin"]
first_cols = [c for c in df.columns if c not in first_cols]
df = df[first_cols + other_cols]
```

```python
df = df.drop(columns=["YY", "MM", "DD", "hh", "mm", "hour"])
```
